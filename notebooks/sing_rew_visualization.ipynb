# Movement Analysis (SLEAP traces)

to the reward-port poke (Hit / False Alarm) or the next trial initiation (Miss / 
Correct Rejection). Multiple dates are overlaid and colored by session (compare early 
vs. late); `show_average` adds a mean (arc-length) trace, `plot_subcategories` splits 
each category into its subcategories.

```python
from hypnose.visualization.movement_analysis.sing_rew_movement import plot_category_traces

# 4 figures (Hit / Miss / FA / CR); show_average adds a mean trace across overlaid sessions. 
# plot_subcategories plots all subcategories (Hit: rewarded_hit, port_error_hit, anticipatory_hit, anticipatory_port_error; miss: omission_miss, forfeit_miss; FA: completed_FA, aborted_FA; CR: passive_CR, active_CR). 
plot_category_traces(59, dates=[20260615], plot_subcategories=False, show_average=True, save=True)
```
